```python
import re, requests
from bs4 import BeautifulSoup
from pprint import pprint
```

```python
vacancy = "Программист C++"
depth = 5
```

```python
headers = {'User-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36'}
proxies = {'http': 'http://user:pwd@192.168.50.26:3128',
           'https': 'https://user:pwd@192.168.50.26:3128',
           'ftp': 'ftp://user:pwd@192.168.50.26:3128' }
```

```python
not_found = True

for i in range(depth):
    link = f'https://hh.ru/search/vacancy?text={vacancy}&page={i}'
    req = requests.get(link, headers=headers, proxies=proxies)
    soup = BeautifulSoup(req.text)
    vac_html = soup.find('div', {'class': 'vacancy-serp'})
    
    for vac in vac_html.findChildren("div" , recursive=False):
        vac_1 = vac.find('a', {'class': 'bloko-link HH-LinkModifier'})
        if vac_1 is None: continue
        vac_name = vac_1.text
        vac_href = re.findall("^(.+)\?", vac_1.get('href'))[0]
        vac_2 = vac.find('div', {'class': 'vacancy-serp-item__compensation'})
        vac_salary = vac_2.text if vac_2 is not None else "по результатам собеседования"
        vac_3 = vac.find('span', {'class': 'vacancy-serp-item__meta-info'})
        vac_city = vac_3.text if vac_3 is not None else ""
        print(f"{vac_name} ({vac_city}) з/п {vac_salary} {vac_href}")
        not_found = False

if not_found: print("Вакансий не найдено")
```
