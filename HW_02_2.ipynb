```python
import json, re, requests
from pprint import pprint

headers = {'User-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36'}
#proxies = {'http': 'http://user:pwd@192.168.50.26:3128',
#           'https': 'https://user:pwd@192.168.50.26:3128',
#           'ftp': 'ftp://user:pwd@192.168.50.26:3128' }

def get_wiki_link(topic):
    link='https://ru.wikipedia.org/wiki/' + topic.capitalize()
    return link

def get_page(url):
    try:
        req = requests.get(url, headers=headers)
        #req = requests.get(url, headers=headers, proxies=proxies)
        req.encoding = req.apparent_encoding
        html = req.text
        #print(req.headers)
        return html
    except:
        return None

def get_url_text(url):
    html_content = get_page(url) 
    if html_content is None: return []
    words = re.findall("[а-яА-ЯёЁ]{3,}", html_content)
    return words

def get_common_words(url):
    words = get_url_text(url)
    rate = {}
    for word in words:
        if word in rate:
            rate[word] += 1
        else:
            rate[word] = 1
    rate_list = list(rate.items())
    rate_list.sort(key=lambda x: -x[1])
    return rate_list

def get_topic_refs(topic):
    url = get_wiki_link(topic)
    html_content = get_page(url)
    urls = re.findall("<li><a rel=\"nofollow\" class=\"external text\" href=\"((https?:\/\/){1}([-\d\w\.]+)([a-z]{2,6}\.?)(\/[-\d\w\.]*)*\/?)\">", 
                       html_content)
    #words = re.findall("(https?:\/\/){1}([-\d\w\.]+)([a-z]{2,6}\.?)(\/[-\d\w\.]*)*\/?", html_content)
    urls = [x[0] for x in urls]
    return urls


refs = get_topic_refs('Меркурий')
for ref in refs:
    print(ref)
    pprint(get_common_words(ref)[:10])
```
